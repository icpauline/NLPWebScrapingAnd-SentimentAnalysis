```python
from requests_html import HTMLSession
import pandas as pd


s = HTMLSession()
r = s.get(url)
# r.html.render(sleep=1)
items = r.html.find('div[data-asin]')

data_asin = []

for item in items:
    if item.attrs['data-asin'] != '':
        data_asin.append(item.attrs['data-asin'])
```

```python
link=[]
cookie = {}
for i in range(len(data_asin)):
  url="https://www.amazon.in/dp/"+data_asin[i]
  response=requests.get(url,cookies= cookie,headers=header)
  if response.status_code==200:
        print(response)
  else:
        print('Error')
  soup=BeautifulSoup(response.content)
  for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
    link.append(i['href'])
```

```python
link
```

```python
len(link)
```

```python
# Now we have the ‘see all review’ links. Using this link along with a page number, we can extract the reviews in any number of pages for all the products and save it into a list called reviews[]
reviews=[]
for j in range(len(link)):
  for k in range(100):
    response=Searchreviews(link[j]+'&pageNumber='+str(k))
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("span",{'data-hook':"review-body"}):
      reviews.append(i.text)
```

```python
len(reviews)
```

```python
rev={'reviews':reviews}
review_data=pd.DataFrame.from_dict(rev)
pd.set_option('max_colwidth',800)
review_data.head(10)
```

### Import 'SentimentIntensityAnalyzer'

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df1['scores'] = df1['reviews'].apply(lambda review: sid.polarity_scores(review))
```

```python
df1.head()
```

```python
df1['compound']  = df1['scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
df1.head()
```

```python
df1['comp_score'] = df1['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
```
